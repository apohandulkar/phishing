```python
import ipaddress
import re
import urllib.request
from bs4 import BeautifulSoup
import socket
import requests
from googlesearch import search
import whois
from datetime import date, datetime
import time
from dateutil.parser import parse as date_parse
from urllib.parse import urlparse


def get_features(df):
    for url in df.url:
        count=0
        def __init__(self,url):
            self.features = ()
            self.url = url
            self.domain = ""
            self.whois_response = ""
            self.urlparse = ""
            self.response = ""
            self.soup = ""

            try:
                self.response = requests.get(url)
                self.soup = BeautifulSoup(response.text, 'html.parser')
            except:
                pass
            try:
                self.urlparse = urlparse(url)
                self.domain = self.urlparse.netloc
            except:
                pass

            try:
                self.whois_response = whois.whois(self.domain)
            except:
                pass


        
        def __init__(self,url):
            self.features = ()
            self.url = url
            self.domain = ""
            self.whois_response = ""
            self.urlparse = ""
            self.response = ""
            self.soup = ""

            try:
                self.response = requests.get(url)
                self.soup = BeautifulSoup(response.text, 'html.parser')
            except:
                pass

            try:
                self.urlparse = urlparse(url)
                self.domain = self.urlparse.netloc
            except:
                pass

            try:
                self.whois_response = whois.whois(self.domain)
            except:
                pass
            
        
         # 1.UsingIp
        def UsingIp(url):
            try:
                ipaddress.ip_address(url)
                return -1
            except:
                return 1
        df['UsingIP']= UsingIp(url)

        # 2.longUrl
        def longUrl(url):
            if len(url) < 54:
                return 1
            if len(url) >= 54 and len(url) <= 75:
                return 0
            return -1
        df['LongURL']=longUrl(url)

        # 3.shortUrl
        def shortUrl(url):
            match = re.search('bit\.ly|goo\.gl|shorte\.st|go2l\.ink|x\.co|ow\.ly|t\.co|tinyurl|tr\.im|is\.gd|cli\.gs|'
                        'yfrog\.com|migre\.me|ff\.im|tiny\.cc|url4\.eu|twit\.ac|su\.pr|twurl\.nl|snipurl\.com|'
                        'short\.to|BudURL\.com|ping\.fm|post\.ly|Just\.as|bkite\.com|snipr\.com|fic\.kr|loopt\.us|'
                        'doiop\.com|short\.ie|kl\.am|wp\.me|rubyurl\.com|om\.ly|to\.ly|bit\.do|t\.co|lnkd\.in|'
                        'db\.tt|qr\.ae|aurl\.ly|goo\.gl|bitly\.com|cur\.lv|tinyurl\.com|ow\.ly|bit\.ly|ity\.im|'
                        'q\.gs|is\.gd|po\.st|bc\.vc|twitthis\.com|u\.to|j\.mp|buzurl\.com|cutt\.us|u\.bb|yourls\.org|'
                        'x\.co|prettylinkpro\.com|scrnch\.me|filoops\.info|vzturl\.com|qr\.net|1url\.com|tweez\.me|v\.gd|tr\.im|link\.zip\.net', url)
            if match:
                return -1
            return 1
        df['ShortURL']=shortUrl(url)


        # 4.Symbol@
        def symbol(url):
            if re.findall("@",url):
                return 1
            return -1
        df['Symbol@']=symbol(url)


        # 5.Redirecting//
        def redirecting(url):
            if url.rfind('//')>6:
                return -1
            return 1
        df['Redirecting//']=redirecting(url)
        
        # 6.prefixSuffix
        def prefixSuffix(url):
            try:
                match = re.findall('\-', url.domain)
                if match:
                    return -1
                return 1
            except:
                return -1
        df['PrefixSuffix-']=prefixSuffix(url)

        # 7.SubDomains
        def SubDomains(url):
            dot_count = len(re.findall("\.", url))
            if dot_count == 1:
                return 1
            elif dot_count == 2:
                return 0
            return -1
        df['SubDomains']=SubDomains(url)

        # 8.HTTPS
        def Hppts(url):
            try:
                https = urlparse.scheme
                if 'https' in https:
                    return 1
                return -1
            except:
                return 1
        df['HTTPS']=Hppts(url)


        # 9. HTTPSDomainURL
        def HTTPSDomainURL(url):
            try:
                if 'https' in url.domain:
                    return -1
                return 1
            except:
                return -1
        df['HTTPSDomainURL']=HTTPSDomainURL(url)

        # 10. RequestURL
        def RequestURL(url):
            try:
                for img in url.soup.find_all('img', src=True):
                    dots = (x.start(0) for x in re.finditer('\.', img('src')))
                    if url in img('src') or url.domain in img('src') or len(dots) == 1:
                        success = success + 1
                    i = i+1

                for audio in url.soup.find_all('audio', src=True):
                    dots = (x.start(0) for x in re.finditer('\.', audio('src')))
                    if url in audio('src') or url.domain in audio('src') or len(dots) == 1:
                        success = success + 1
                    i = i+1

                for embed in url.soup.find_all('embed', src=True):
                    dots = (x.start(0) for x in re.finditer('\.', embed('src')))
                    if url in embed('src') or url.domain in embed('src') or len(dots) == 1:
                        success = success + 1
                    i = i+1

                for iframe in url.soup.find_all('iframe', src=True):
                    dots = (x.start(0) for x in re.finditer('\.', iframe('src')))
                    if url in iframe('src') or url.domain in iframe('src') or len(dots) == 1:
                        success = success + 1
                    i = i+1

                try:
                    percentage = success/float(i) * 100
                    if percentage < 22.0:
                        return 1
                    elif((percentage >= 22.0) and (percentage < 61.0)):
                        return 0
                    else:
                        return -1
                except:
                    return 0
            except:
                return -1
            
        df['RequestURL']=RequestURL(url)

        # 11. AnchorURL
        def AnchorURL(url):
            try:
                i,unsafe = 0,0
                for a in url.soup.find_all('a', href=True):
                    if "#" in a('href') or "javascript" in a('href').lower() or "mailto" in a('href').lower() or not (url in a('href') or url.domain in a('href')):
                        unsafe = unsafe + 1
                    i = i + 1

                try:
                    percentage = unsafe / float(i) * 100
                    if percentage < 31.0:
                        return 1
                    elif ((percentage >= 31.0) and (percentage < 67.0)):
                        return 0
                    else:
                        return -1
                except:
                    return -1

            except:
                return -1
            
        df['AnchorURL']=AnchorURL(url)




        # 12. InfoEmail
        def InfoEmail(url):
            try:
                if re.findall(r"(mail\(\)|mailto:?)", url.soap):
                    return -1
                else:
                    return 1
            except:
                return -1
            
        df['InfoEmail']=InfoEmail(url)

        # 13. AbnormalURL
        def AbnormalURL(url):
            try:
                if url.response.text == url.whois_response:
                    return 1
                else:
                    return -1
            except:
                return -1
        df['AbnormalURL']=AbnormalURL(url)

        # 14. WebsiteForwarding
        def WebsiteForwarding(url):
            try:
                if len(url.response.history) <= 1:
                    return 1
                elif len(url.response.history) <= 4:
                    return 0
                else:
                    return -1
            except:
                 return -1
        df['WebsiteForwarding']=WebsiteForwarding(url)



        # 15. DisableRightClick
        def DisableRightClick(url):
            try:
                if re.findall(r"event.button ?== ?2", url.response.text):
                    return 1
                else:
                    return -1
            except:
                 return -1
                
        df['DisableRightClick']=DisableRightClick(url)

        # 16. UsingPopupWindow
        def UsingPopupWindow(url):
            try:
                if re.findall(r"alert\(", url.response.text):
                    return 1
                else:
                    return -1
            except:
                 return -1
                
        df['UsingPopupWindow']=UsingPopupWindow(url)
        

        # 17. IframeRedirection
        def IframeRedirection(url):
            try:
                if re.findall(r"(<iframe>|<frameBorder>)", url.response.text):
                    return 1
                else:
                    return -1
            except:
                 return -1
        df['IframeRedirection']=IframeRedirection(url)
        

        # 18. AgeofDomain
        def AgeofDomain(url):
            try:
                creation_date = url.whois_response.creation_date
                try:
                    if(len(creation_date)):
                        creation_date = creation_date(0)
                except:
                    pass

                today  = date.today()
                age = (today.year-creation_date.year)*12+(today.month-creation_date.month)
                if age >=6:
                    return 1
                return -1
            except:
                return -1
        df['AgeofDomain']=AgeofDomain(url)
        

        # 19. DNSRecording    
        def DNSRecording(url):
            try:
                creation_date = url.whois_response.creation_date
                try:
                    if(len(creation_date)):
                        creation_date = creation_date(0)
                except:
                    pass

                today  = date.today()
                age = (today.year-creation_date.year)*12+(today.month-creation_date.month)
                if age >=6:
                    return 1
                return -1
            except:
                return -1
        df['DNSRecording']=DNSRecording(url)
        

        # 20. WebsiteTraffic   
        def WebsiteTraffic(url):
            try:
                rank = BeautifulSoup(urllib.request.urlopen("http://data.alexa.com/data?cli=10&dat=s&url=" + url).read(), "xml").find("REACH")('RANK')
                if (int(rank) < 100000):
                    return 1
                return 0
            except :
                return -1
        df['WebsiteTraffic']=WebsiteTraffic(url)

        # 21. PageRank
        def PageRank(url):
            try:
                prank_checker_response = requests.post("https://www.checkpagerank.net/index.php", {"name": url.domain})

                global_rank = int(re.findall(r"Global Rank: ((0-9)+)", rank_checker_response.text)(0))
                if global_rank > 0 and global_rank < 100000:
                    return 1
                return -1
            except:
                return -1
        df['PageRank'] = PageRank(url)


        # 22. GoogleIndex
        def GoogleIndex(url):
            try:
                site = search(url, 5)
                if site:
                    return 1
                else:
                    return -1
            except:
                return 1
        df['GoogleIndex']=GoogleIndex(url)

        # 23. LinksPointingToPage
        def LinksPointingToPage(url):
            try:
                number_of_links = len(re.findall(r"<a href=", url.response.text))
                if number_of_links == 0:
                    return 1
                elif number_of_links <= 2:
                    return 0
                else:
                    return -1
            except:
                return -1
        df['LinksPointingToPage']=LinksPointingToPage(url)
```
